## Задание 3

Прочитать данные `https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json`.

Выполнить сохранение в формат CSV. Прочитать полученный файл. Сравнить с исходным.


````python
import pandas as pd

artists_url = "https://raw.githubusercontent.com/yakushinav/omo/main/data/artists.json"

artists_data = pd.read_json(artists_url)

csv_file_path = "artists.csv"
artists_data.to_csv(csv_file_path, index=False)

artists_from_csv = pd.read_csv(csv_file_path)

comparison_result = artists_data.equals(artists_from_csv)

print(comparison_result)
````
